```python
import tqdm
import pandas as pd
import sentence_transformers

# Load the dataset
df = pd.read_csv("/kaggle/input/semantic/cleanedbygks.csv")
df=df.sample(n=10000)
# Create a list of all the product names
product_names = df["TITLE"].tolist()

# Load a pre-trained sentence embedding model
model = sentence_transformers.SentenceTransformer("bert-base-uncased")
```

```python
# Create a dictionary mapping product names to their descriptions
df["PRODUCT_LENGTH"] = df["PRODUCT_LENGTH"].astype(int)
df["DESCRIPTION"] = df["DESCRIPTION"].astype(str)

descriptions = {}
for product_name in tqdm.tqdm(product_names):
    # Get the description of the product
    description = df[df["TITLE"] == product_name]["DESCRIPTION"].values[0]

    # Encode the description
    product_embedding = model.encode(description)

    # Add the product name and its embedding to the dictionary
    descriptions[product_name] = product_embedding
```

```python
# Define a function to search for a product
# product_name = product_name.reshape(1, -1)

# def search(product_name):
#     # Get the embedding of the product name
# #     product_embedding = model.encode(product_name) #ikkada petu ra kindha laga
#     product_embedding = model.encode(product_name).reshape(1, -1)

#     # Find the products that are most semantically similar to the product name
#     similar_products = []
#     for product_name, product_embedding in descriptions.items():
#         similarity = cosine_similarity(product_embedding, product_embedding)
#         if similarity > 0.5:
#             similar_products.append((product_name, similarity))

#     # Sort the products by similarity
#     similar_products.sort(key=lambda x: x[1], reverse=True)

#     # Return the most similar products
#     return similar_products

from sklearn.metrics.pairwise import cosine_similarity

def search(product_name):
    # Get the embedding of the product name
    product_embedding = model.encode(product_name).reshape(1, -1)

    # Find the products that are most semantically similar to the product name
    similar_products = []
    for name, embedding in descriptions.items():
        similarity = cosine_similarity(product_embedding, embedding.reshape(1, -1))
        if similarity > 0.794:
            similar_products.append((name, similarity[0, 0]))

    # Sort the products by similarity
    similar_products.sort(key=lambda x: x[1], reverse=True)

    # Return the most similar products
    return similar_products,len(similar_products)


# Test the search function
print(search("Drawing book"))
```
